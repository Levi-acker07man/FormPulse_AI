```python
import os
import numpy as np
import pandas as pd
import joblib
import matplotlib.pyplot as plt
from google.colab import files
import tensorflow as tf
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, Bidirectional
```

```python
players_df = pd.read_csv('/content/players.csv')
games_details_df = pd.read_csv('/content/games_details.csv', low_memory=False)
games_df = pd.read_csv('/content/games.csv')
```

```python
TARGET_PLAYERS = [
    "Stephen Curry",
    "LeBron James",
    "Kevin Durant",
    "Luka Doncic",
    "Giannis Antetokounmpo"
]
```

```python
cohort_ids = players_df[players_df['PLAYER_NAME'].isin(TARGET_PLAYERS)]['PLAYER_ID'].tolist()
```

```python
player_stats = games_details_df[games_details_df['PLAYER_ID'].isin(cohort_ids)].copy()
```

```python
merged_df = pd.merge(
    player_stats,
    games_df[['GAME_ID', 'GAME_DATE_EST']],
    on='GAME_ID',
    how='inner'
)
```

```python
merged_df['GAME_DATE_EST'] = pd.to_datetime(merged_df['GAME_DATE_EST'])
processed_df = merged_df.sort_values(by='GAME_DATE_EST').reset_index(drop=True)
```

```python
def parse_minutes(min_str):
    if pd.isna(min_str): return 0.0
    min_str_clean = str(min_str).strip()
    if ':' in min_str_clean:
        parts = min_str_clean.split(':')
        try: return float(parts[0]) + (float(parts[1]) / 60.0)
        except: return 0.0
    try: return float(min_str_clean)
    except: return 0.0

processed_df['MIN_FLOAT'] = processed_df['MIN'].apply(parse_minutes)
processed_df['DAYS_REST'] = processed_df.groupby('PLAYER_ID')['GAME_DATE_EST'].diff().dt.days
processed_df['DAYS_REST'] = processed_df['DAYS_REST'].fillna(3.0)

processed_df = processed_df.dropna(subset=['PTS', 'FGA', 'FTA'])
feature_columns = ['PTS', 'MIN_FLOAT', 'DAYS_REST', 'FGA', 'FTA']
```

```python
scaler = MinMaxScaler(feature_range=(0, 1))
processed_df[feature_columns] = scaler.fit_transform(processed_df[feature_columns])
```

```python
WINDOW_SIZE = 5
X_seq, y_target = [], []
```

```python
##critical:build sequence for the cohert
for player_id, group in processed_df.groupby('PLAYER_ID'):
    player_matrix = group[feature_columns].values

    # Skip if they don't have enough games to form a sequence
    if len(player_matrix) <= WINDOW_SIZE:
        continue

    for i in range(WINDOW_SIZE, len(player_matrix)):
        X_seq.append(player_matrix[i-WINDOW_SIZE:i])
        y_target.append(player_matrix[i, 0])
```

```python
X = np.array(X_seq)
y = np.array(y_target)
```

```python
##test train/split
split_index = int(0.85 * len(X))
X_train, X_test = X[:split_index], X[split_index:]
y_train, y_test = y[:split_index], y[split_index:]
```

```python
model = Sequential([
    Bidirectional(LSTM(units=64, return_sequences=True), input_shape=(WINDOW_SIZE, len(feature_columns))),
    Dropout(0.2),
    LSTM(units=32, return_sequences=False),
    Dropout(0.2),
    Dense(units=16, activation='relu'),
    Dense(units=1)
])
```

```python
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss='mean_squared_error',
    metrics=['mae']
)
```

```python
history = model.fit(
    X_train, y_train,
    epochs=65,
    batch_size=16,
    validation_data=(X_test, y_test),
    verbose=1
)
```

```python
plt.figure(figsize=(10, 4))
plt.plot(history.history['loss'], label='Training Loss (MSE)')
plt.plot(history.history['val_loss'], label='Validation Loss (MSE)')
plt.title(f'FormPulse AI - Generalized Cohort Optimization')
plt.xlabel('Epoch Cycles')
plt.ylabel('Loss Value')
plt.legend()
plt.show()
```

```python
print("Exporting Engine and Scaler for the Interactive App...")
model.save('formpulse_model.keras')
joblib.dump(scaler, 'scaler.pkl')

files.download('formpulse_model.keras')
files.download('scaler.pkl')
```
